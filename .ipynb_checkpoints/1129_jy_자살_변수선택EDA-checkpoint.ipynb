```python
import pandas as pd
import matplotlib.pyplot as plt
import missingno as msno
import numpy  as np
import seaborn as sns
%matplotlib inline

path = './data/'
path바탕화면 = 'C:/Users/COM/Desktop/자살통계/'
pd.options.display.max_rows = 150
pd.options.display.max_columns = 350
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
연령별자살 = pd.read_csv(path바탕화면+'사망원인_236항목__성_연령_5세_별_사망자수__사망률_20191129154932.csv', encoding='cp949')
```

```python
연령별자살
```

```python
#자살사망률이 높은 연령대와 남녀차이
연령별자살사망률 = 연령별자살.loc[연령별자살['사망원인별(236항목)']=='고의적 자해(자살) (X60-X84)']
연령별자살사망률 = 연령별자살사망률.groupby(['연령(5세)별','성별']).sum()
연령별자살사망률 = 연령별자살사망률.reset_index()
연령별자살사망률
```
