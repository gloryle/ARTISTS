```python
import pandas as pd
import matplotlib.pyplot as plt
import missingno as msno
import numpy  as np
import seaborn as sns
%matplotlib inline

path = './data/'
pd.options.display.max_rows = 150
pd.options.display.max_columns = 350
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv(path + '변수중요도_감염병.csv',encoding='cp949')
sido = pd.read_csv(path+'sido_imp_10.csv', encoding='cp949')
```

```python
sido_y = sido.iloc[:,-7:]
sido_x_scaled = sido.iloc[:,3:-7]
sido_name = sido.iloc[:,:3]
```

# 변수선택

```python
#RFE로 소팅해
variables_df = df.sort_values('RFE', ascending=False)
#RFE가 유의미한 변수라고 판단한 변수의 이름을 가져온다.
variables_name_20 = variables_df.loc[variables_df.RFE==True,'feature'].values
#RFE가 선택한 독립변수 데이터프레임
variables_20 = sido_x_scaled.loc[:,variables_name_20]
#variables_20
```

```python
variables_df = df.sort_values('Kbest', ascending=False)
#RFE가 유의미한 변수라고 판단한 변수의 이름을 가져온다.
variables_name_20 = variables_df.loc[variables_df.Kbest==True,'feature'].values
#RFE가 선택한 독립변수 데이터프레임
variables_20 = sido_x_scaled.loc[:,variables_name_20]
#variables_20
```

```python
#RFE로 소팅해
variables_df_2 = df.sort_values('RFE', ascending=False)
#RFE가 유의미한 변수라고 판단한 변수의 이름을 가져온다.
variables_name_20_2 = variables_df_2.loc[variables_df_2.RFE==True,'feature'].values
#RFE가 선택한 독립변수 데이터프레임
variables_20_2 = sido_x_scaled.loc[:,variables_name_20_2]
#variables_20
```

# 변수선택방법별 모델비교

### 전체변수넣고

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
mlp = MLPClassifier(hidden_layer_sizes=(50,30))
mlp.fit(sido_x_scaled, sido.감염병)
pred = mlp.predict(sido_x_scaled)
```

```python
def KFold_meanscore(ary):
    tot = 0
    for i in ary:
        tot += i
    return tot/len(ary)
```

```python
#crossvalscore이용
cv = KFold(3, shuffle=True, random_state=0)
ary = cross_val_score(mlp, sido_x_scaled, sido.감염병, cv=cv, scoring='f1')
ary, KFold_meanscore(ary)
```

```python
#분류메트릭이용 - true와 pred를 넣는거니까.
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
accuracy_score(sido.감염병,pred), precision_score(sido.감염병, pred), recall_score(sido.감염병, pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(sido.감염병,pred, target_names=['123','45']))
```
