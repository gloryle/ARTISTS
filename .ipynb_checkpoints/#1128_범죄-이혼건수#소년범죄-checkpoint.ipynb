## 중앙선그리는 함수 선언
scatter(x="이혼건수", y="5대범죄 발생건수", color="crime", df=sigungu_nxy, medianline=1, text=1, palette='picnic', hovername='지역')

```python
import plotly.express as px
import plotly.graph_objects as go
def scatter(x, y, color, df, hovername=None,text=False, medianline=False, palette='fall'):
    if text == True:
        fig = px.scatter(df, x=x, y=y, text='시군구',
                         hover_name=hovername, color=color, color_continuous_scale=palette
                        )
        fig.update_traces(textposition='top center')
    else:
        fig = px.scatter(df, x=x, y=y,
                         hover_name=hovername, color=color, color_continuous_scale=palette
                        )
        fig.update_traces(textposition='top center')
    if medianline==True:
        #x축 중앙값 그리기
        fig.add_trace(
            go.Scatter(x=[df[x].median()]*2,
                       y=[-3000,3000],
                       name="median x: "+str(round(df[x].median(),2)),
                       mode="lines",
                       line = dict(color = ('rgb(0,0,0)'),width=0.5),
                       visible=True,
                      ))

    #y축 중앙값 그리기
        fig.add_trace(
            go.Scatter(x=[-3000,3000],
                       y=[df[y].median()]*2,
                       name="median y: "+str(round(df[y].median(),2)),
                       mode="lines",
                       line = dict(color = ('rgb(0,0,0)'),width=0.5),
                       visible=True
                      ))
        fig.update_layout(legend=dict(x=-.05, y=1.15))

    #플롯 레이아웃(x범위, y범위) 설정
    fig.update_xaxes(range=[df[x].min()*0.8, df[x].max()*1.05]) #min값 음수인지 아닌지 확인 후 범위 조절
    fig.update_yaxes(range=[df[y].min()*0.5, df[y].max()*1.05]) #min값 음수인지 아닌지 확인 후 범위 조절

    fig.show()
```

### 외부데이터로 추가정보 만들기

```python
#시군구별 청소년인구 데이터를 가져옴 : popul
popul = pd.read_csv(path+'행정구역_시군구_별_1세별_주민등록인구_20191125181624.csv', encoding='cp949')
popul = popul.loc[popul['행정구역(시군구)별(2)']!='소계']
a = popul.groupby(['행정구역(시군구)별(1)','행정구역(시군구)별(2)']).sum()
a.reset_index(inplace=True)
a.columns = ['시도','시군구','청소년인구', '남자인구수 (명)', '여자인구수 (명)']
sigungu_youthpopul = a.loc[:,['시도','시군구','청소년인구']]
```

```python
#시군구별 소년범죄 건수 데이터를 가져옴 : youthcrime
youthcrime = pd.read_csv(path+'청소년범죄발생건수.csv', encoding='cp949')
youthcrime.replace({'강원':'강원도','경기':'경기도','경남':'경상남도','경북':'경상북도',
                   '광주':'광주광역시','대구':'대구광역시','대전':'대전광역시','부산':'부산광역시',
                   '서울':'서울특별시','울산':'울산광역시','인천':'인천광역시',
                   '전남':'전라남도','전북':'전라북도','충남':'충청남도','충북':'충청북도','제주':'제주특별자치도'},inplace=True)
sigungu_youthcrime = youthcrime.loc[:,['시도','시군구','합계']]
sigungu_youthcrime.columns=['시도','시군구','소년범죄건수']
youthcrime
sigungu_youthcrime
```

```python
#시군구별 전체범죄 건수 데이터를 가져옴 : totalcrime
totalcrime = pd.read_csv(path+'전체범죄발생건수.csv', encoding='cp949')
totalcrime.replace({'강원':'강원도','경기':'경기도','경남':'경상남도','경북':'경상북도',
                   '광주':'광주광역시','대구':'대구광역시','대전':'대전광역시','부산':'부산광역시',
                   '서울':'서울특별시','울산':'울산광역시','인천':'인천광역시',
                   '전남':'전라남도','전북':'전라북도','충남':'충청남도','충북':'충청북도','제주':'제주특별자치도'},inplace=True)
sigungu_totalcrime = totalcrime.loc[:,['시도','시군구','합계']]
sigungu_totalcrime.columns=['시도','시군구','전체범죄건수']
sigungu_totalcrime
```

```python
# 소년범죄비율을 구한 데이터프레임 : df (위 3개의 외부데이터를 병합해 만듬)
df = sigungu_youthpopul.merge(sigungu_youthcrime).merge(sigungu_totalcrime)
df['소년범죄비율']=df.소년범죄건수/df.전체범죄건수
df
```

### 군분류
<pre>
1사분면에 존재하는 주황색점들만 group1으로 지정해 이 그룹의 경향성을 살펴보고자 한다.
가설: 이혼건수가 높고 범죄발생률이 높은 지역은 --> group1으로 따로 지정해
      전체범죄중에서 소년범죄의 비율이 높을 것이다. --> 가설확인
</pre>

```python
#군분류: 1사분면: 이혼건수와 범죄발생건수가 중앙값보다 높은
cond=(sigungu_nxy.이혼건수>sigungu_nxy.이혼건수.median())&(sigungu_nxy['5대범죄 발생건수']>sigungu_nxy['5대범죄 발생건수'].median())
df['group1'] = 0
df.loc[cond&(sigungu_nxy.crime==1),'group1']=1
len(df.loc[df.group1==1])
```

```python
#이혼건수
sigungu_divorce = sigungu_nxy.loc[:,['시도','시군구','이혼건수','crime']]
ultimatedivorce = df.merge(sigungu_divorce)
len(ultimatedivorce)
```
