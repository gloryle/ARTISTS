## 데이터가 차있는 수와 비율을 확인
<pre>
sido_nullax1: 전체변수에 대해 몇개의 데이터가 채워져있는가. 그 비율은?
sido_nullax0: 전체시도에 대해 몇개의 데이터가 채워져있는가. 그 비율은?
</pre>

```python
sido_nullax1 = pd.DataFrame(data = sido_x.notnull().sum(axis=0), columns={'data'} )
sido_nullax1['데이터비율']=round(sido_nullax1.data/(len(sido_x)),2)
# sido_nullax1.reset_index(inplace=True)
sido_nullax1
#모든 시도에 대해 NaN인 열은 삭제한다.data가 0인경우 dropna(how='all')
#80%이상 결측인 열은 삭제한다.
#이전 자료로 결측을 채울 수 있는지 확인한다.
```

```python
df = pd.DataFrame(data = sido_x.notnull().sum(axis=1), columns={'data'} )
sido_nullax0 = pd.concat([sido_name.시도, df], axis=1)
sido_nullax0['데이터비율']=round(sido_nullax0.data/sido_x.shape[1],2)
sido_nullax0
```

## 열의 분류와 처리

```python
#아무데이터도 없는 열 : 바로 삭제
len(sido_nullax1.loc[(sido_nullax1.데이터비율==0)])
```

```python
#데이터가 모두 완전한 열: 바로 변수
len(sido_nullax1.loc[(sido_nullax1.데이터비율==1)])
```

```python
#데이터가 일부만 차있는 열: 80%이상 있는 열에 대해 결측치 처리 진행
len(sido_nullax1.loc[(sido_nullax1.데이터비율!=0)&(sido_nullax1.데이터비율!=1)])
```
