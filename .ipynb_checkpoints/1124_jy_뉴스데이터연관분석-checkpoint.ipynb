```python
뉴스크롤링 = pd.read_csv(path + f'{keyword} 웹크롤링.csv', encoding='cp949')
뉴스크롤링
n_lst=[]
for i in range(len(뉴스크롤링)):
    n_lst.append(뉴스크롤링.title[i])
```

```python
n_lst
```

```python
n_lst[0]
```

```python
','.join(n_lst[0])
```

```python
from konlpy.tag import Komoran
komoran = Komoran()
komoran.nouns(n_lst[0])
```

```python
lst=[]
lst2=[3,4,5]
lst3=[6,5,4]
#1번방법
lst.append(lst2)
lst.append(lst3)
#2번방법
[lst2,lst3]
```

```python
transactions=[]
for i in n_lst:
    lst = komoran.nouns(i)
    transactions.append(lst)
```

```python
transactions
```

```python
from apyori import apriori
rules = apriori(transactions, min_support=0.1, min_confidence=0.1)
#지지도가 너무 작으면 연관분석에 많은 시간이 소요된다.
results= list(rules)
```

```python
results
```

```python
result_df = pd.DataFrame(None, columns=['lhs','rhs','support','confidence','lift'])
index=0
for result in results:
    support = result[1]
    ordered_statistic = result[2]
    for item in ordered_statistic:
        lhs = ' '.join(x.strip() for x in item[0])
        rhs = ' '.join(x.strip() for x in item[1])
        confidence = item[2]
        lift = item[3]
        result_df.loc[index] = [lhs, rhs, support, confidence, lift]
        index +=1
```
