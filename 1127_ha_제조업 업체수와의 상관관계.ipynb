### '제조업 업체수'와 변수별 상관관계

```python
sigungu_2015.drop(['년도','시도', '시군구', '지역'], axis=1, inplace=True)
sigungu_2015
```

```python
data = sigungu_2015.corr(method='pearson')['제조업 업체수'].sort_values(ascending=False).to_frame()
data
```

```python
plt.figure(figsize = (1,50))

y='제조업 업체수'

sns.heatmap(data = data, annot=True, fmt = '.2f', linewidths=.5, cmap='Reds')
plt.title(y, fontsize=15)
    
plt.show()
```
