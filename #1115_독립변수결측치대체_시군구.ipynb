```python
def info_df(data):
    info_df = pd.DataFrame({"type":data.dtypes,
                            'null_count':data.isnull().sum(),
                           'rate':data.isnull().sum()/data.isnull().count() * 100}).T
    return info_df
```

### 평균치 채워주는 함수

```python
def mean_score(columns,*r):
    
    a = 0
    b = len(r)
    for reg in r:
        a += (sigungu.loc[sigungu["지역"]==reg,columns].values)
    
    c= a/b
    
    return c
```

### 결측치 채워주는 함수

```python
sigungu.loc[sigungu["지역"]=='2015_대구광역시_남구','하천연장'].values
```

```python
def fill_score(columns,score,*r):
    
    
    for reg in r:
        sigungu.loc[sigungu["지역"]==reg,columns] = sigungu.loc[sigungu["지역"]==reg,columns].fillna(score)
```
