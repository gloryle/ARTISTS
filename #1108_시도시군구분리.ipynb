```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import missingno as msno
path = './data/'
data = pd.read_csv(path + 'A1_년도별_지역_안전지표.csv', encoding='cp949', engine='python')

plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
msno.matrix(data)
```

```python
sido = data.loc[data.시군구.isnull()==True]
sido.to_csv(path+'sido.csv',encoding='cp949',index=False)
print(sido.shape)
msno.matrix(sido)
```

```python
sigungu = data.loc[data.시군구.isnull()==False]
sigungu.to_csv(path+'sigungu.csv',encoding='cp949',index=False)
sigungu.shape
print(sigungu.shape)
msno.matrix(sigungu)
```
