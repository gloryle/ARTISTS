```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy  as np

import matplotlib.pyplot as plt
from pyvis.network import Network

pd.options.display.max_rows = 150
pd.options.display.max_columns = 350
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

path = './data/'
```

```python
sigungu = pd.read_csv(path+'imp_시군구별지역안전지표_2015.csv', encoding='cp949', engine='python')
```

```python
#독립변수만 뽑기
sigungu = sigungu.iloc[:,4:-7]
sigungu
```

```python
#독립변수간 상관계수
links = sigungu.corr().stack().reset_index()
links.columns = ['from', 'to','value']
links
```

```python
#상관계수 높은 것들 필터링(본인과의 상관관계는 제외)
links_filtered=links.loc[ (abs(links['value']) > 0.6) & (links['from'] != links['to']) ]
links_filtered
```

```python
#https://pyvis.readthedocs.io/en/latest/tutorial.html 참고

got_net = Network(height="750px", width="100%",font_color="black")

# set the physics layout of the network
got_net.barnes_hut()
got_data = links_filtered
sources = got_data['from']
targets = got_data['to']
weights = got_data['value']

edge_data = zip(sources, targets, weights)

for e in edge_data:
    src = e[0]
    dst = e[1]
    w = e[2]

    got_net.add_node(src, src, title=src)
    got_net.add_node(dst, dst, title=dst)
    got_net.add_edge(src, dst, value=w)

neighbor_map = got_net.get_adj_list()

# add neighbor data to node hover data
for node in got_net.nodes:
    node["title"] += " Neighbors:<br>" + "<br>".join(neighbor_map[node["id"]])
    node["value"] = len(neighbor_map[node["id"]])

got_net.show("gameofthrones.html")
```
