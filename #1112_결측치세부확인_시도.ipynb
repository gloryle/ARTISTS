```python
column = ''
def find_nan(column=column, sido_w=sido_w):
    df = sido_w.loc[(sido_w.년도==2015),['년도','시도','지역',column]]
    return df
def find_trend(sido, column=column, sido_w=sido_w):
    df2 = sido_w.loc[sido_w.시도==sido,['년도','시도','지역',column]]
    return df2
def draw_graph(column=column):
    fig,ax = plt.subplots(ncols=2,figsize=(8,2))
    sns.distplot(find_nan(column)[column].dropna(), ax=ax[0])
    sns.boxplot(x=find_nan(column)[column], data=find_nan(column), ax=ax[1])
def draw_graph_trend(sido, column=column):
    fig,ax = plt.subplots(ncols=2,figsize=(8,2))
    sns.distplot(find_trend(sido,column)[column].dropna(), ax=ax[0])
    sns.boxplot(x=find_trend(sido,column)[column], data=find_trend(sido,column), ax=ax[1])
```

```python
#주변대기오염도 계산
def mean_score3(r1,r2,r3,columns, sido_wide=sido_w):
    a=((sido_wide.loc[sido_wide["지역"]==r1,columns].values)
     +(sido_wide.loc[sido_wide["지역"]==r2,columns].values)
     +(sido_wide.loc[sido_wide["지역"]==r3,columns].values))/3
    return a
mean_score3('2015_충청남도','2015_충청북도','2015_대전광역시','대기오염도')
```

```python
def mean_score4(r1,r2,r3,r4,columns, sido_wide=sido_w):
    a=((sido_wide.loc[sido_wide["지역"]==r1,columns].values)
     +(sido_wide.loc[sido_wide["지역"]==r2,columns].values)
     +(sido_wide.loc[sido_wide["지역"]==r3,columns].values)
     +(sido_wide.loc[sido_wide["지역"]==r4,columns].values))/4
    return float(a)
```

```python
def mean_score2(r1,r2,columns,sido_wide=sido_w):
    a=((sido_wide.loc[sido_wide["지역"]==r1,columns].values)
     +(sido_wide.loc[sido_wide["지역"]==r2,columns].values))/2
    return a
```
