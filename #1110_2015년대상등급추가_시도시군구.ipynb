## [시도] : 등급추가

```python
import pandas as pd
path = './data/'
```

```python
sido_wide = pd.read_csv(path+'sido_wide.csv', encoding='cp949', engine='python')
sido_y = pd.read_csv(path + '시도등급.csv', encoding='cp949', engine='python')
```

```python
sido_y.shape
```

```python
sido15 = sido_wide.loc[sido_wide.년도==2015]
sido15.shape
```

```python
sido15.head(1)
```

```python
sido_y.head()
```

```python
sido15_y = sido15.merge(sido_y)
sido15_y
```

```python
sido15_y.to_csv(path+'시도별지역안전지표_2015.csv', encoding='cp949', index=False)
```
